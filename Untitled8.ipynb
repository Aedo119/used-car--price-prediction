checking and treating null values

```python
df.isnull().sum()*100/len(df)
```

```python
#treating numerical colums with propotional imputation
numeric_columns = ['year', 'odometer']
for column in numeric_columns:
    missing_count = df[column].isna().sum()
    if missing_count > 0:
        sampled_values = np.random.choice(df[column].dropna(), size=missing_count, replace=True)
        df.loc[df[column].isna(), column] = sampled_values
```

```python
#treating categorical colums with propotional imputation
columns_to_clean = ['manufacturer', 'model', 'condition', 'cylinders','title_status', 'fuel', 'transmission', 'drive', 'type', 'paint_color']

for column in columns_to_clean:
    if column in df.columns:
        missing_count = df[column].isna().sum()
        if missing_count > 0:
            sampled_values = np.random.choice(
                df[column].dropna(), 
                size=missing_count, 
                replace=True
            )
            df.loc[df[column].isna(), column] = sampled_values
```

```python
#checking again
df.isnull().sum()
```

```python
#checking values of cylinder column 
df['cylinders'].unique()
```

```python
#mapping equivalent numerical values to the column
mapping = {'3 cylinders': 3,'4 cylinders': 4,'5 cylinders': 5,'6 cylinders': 6,'8 cylinders': 8,'10 cylinders': 10,'12 cylinders': 12,'other': np.nan}

df['cylinders'] = df['cylinders'].replace(mapping)
```

```python
# Get value distribution of existing cylinders
missing_mask = df['cylinders'].isna()
cylinder_dist = df['cylinders'].value_counts(normalize=True)
imputed_values = np.random.choice(cylinder_dist.index,size=missing_mask.sum(), p=cylinder_dist.values)

# Fill missing values proportionally
df.loc[missing_mask, 'cylinders'] = imputed_values
```

Checking for kurtosis and skewness

```python
import seaborn as sns
from scipy.stats import skew, kurtosis

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(15, len(numeric_cols) * 4))

for i, col in enumerate(numeric_cols, 1):
    plt.subplot(len(numeric_cols), 1, i)
    sns.kdeplot(df[col].dropna(), fill=True)
    sk = skew(df[col].dropna())
    kurt = kurtosis(df[col].dropna())
    plt.title(f'{col}\nSkewness: {sk:.2f}, Kurtosis: {kurt:.2f}')
    plt.xlabel(col)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

removing outliers in this column

```python
def remove_outliers_brandwise(df, col, group_col):
    cleaned_df = pd.DataFrame()
    for brand, sub_df in df.groupby(group_col):
        Q1 = sub_df[col].quantile(0.25)
        Q3 = sub_df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        sub_clean = sub_df[(sub_df[col] >= lower_bound) & (sub_df[col] <= upper_bound)]
        cleaned_df = pd.concat([cleaned_df, sub_clean])
    return cleaned_df
df1 = remove_outliers_brandwise(df, 'price', 'manufacturer')
```

Comparing the performance of each model

```python
import pandas as pd
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Dictionary to store results
results = []

models = {
    "Linear Regression": lr,
    "Ridge": grid_ridge.best_estimator_ if 'grid_ridge' in locals() else ridge,
    "Lasso": lasso,
    "ElasticNet": elastic,
    "Decision Tree": dt
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results.append({
        'Model': name,
        'MSE': mse,
        'RMSE': rmse,
        'MAE': mae,
        'R²': r2
    })

results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by='R²', ascending=False)
print(results_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.barplot(x='R²', y='Model', data=results_df, palette="viridis")
plt.title("Model Comparison by R²")
plt.xlim(0, 1)
plt.show()
```
